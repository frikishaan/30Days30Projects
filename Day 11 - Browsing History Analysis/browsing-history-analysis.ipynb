# Browsing history analyzer &#128187;

The notebook helps you find how many times you had visited a particular websites during a given period of time. In order to use this notebook you have to download your data for chrome browing history from [Google](http://takeout.google.com/) and replace the file name &#128071; with the file name of the data you have downloaded in **json** format.

####  Please make sure you have required library installed.

###### Packages required - 
 - jupyter notebook
 - pandas
 - urllib 
 - matplotlib

```python
# importing pandas library

import pandas as pd
from pandas.io.json import json_normalize
```

### Import data file 

```python
# Import data

d = pd.read_json('<your-filename.json>') # Replace name with the filename
d.head()
```

```python
# Normalize data

data = json_normalize(d['Browser History']) 
data.tail(5) 
```

### Drop unnecessary columns

```python
urls = data.drop(['client_id', 'favicon_url', 'page_transition', 'time_usec', 'title'], axis=1)
urls.head(5)
```

### Parse the hostname from the urls using urllib

```python
from urllib.parse import urlparse
```

```python
# Replace each url to the corresponding hostname using Pandas apply() function

urls = urls['url'].apply(lambda x: urlparse(x).netloc)
```

```python
# Prints unique values i.e. unique websites you have visited so far

uniques = urls.unique()
len(uniques)
```

### Count the frequencies of each website using collections library

```python
import collections

# frequencies
frequencies = collections.Counter(urls)

# Sort the urls with frequencies as a key
websites = sorted(frequencies, key=frequencies.get, reverse=True)

# Sort the Visits count
visits = sorted(frequencies.values(), reverse=True)

# Print result
for i,j in zip(websites, visits):
    print('{} - {}'.format(i, j))
```

#### For above lines you'll see an output as a sorted list like- 

youtube.com - 13292

github.com - 10046

udemy.com - 8745

google.com - 7564

### Display result using a bar chart using matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plot a bar chart with first 20 values representing 20 most visited domains/websites by you for a given time period of time

```python
plt.figure(figsize=(20,10))
plt.bar(webistes[:20],visits[:20], color="#F97F51")
plt.xticks(rotation='vertical')
plt.xlabel('Websites')
plt.ylabel('Number of Visits')
plt.show()
```
